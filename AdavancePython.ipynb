ATM PROGRAM WITH PIN

```python
balance =  0
transaction = []
pin = 1609
cnt = 0
approve = False
while cnt <3:
  print(f"Available chances :{3-cnt}")
  code = int(input("Enter your pin :"))
  if code==pin:
    approve = True
    break
  else:
    print("Wrong Pin")
    cnt+=1
while(approve):
  print("\n1 for Deposit")
  print("2 for withdraw")
  print("3 for Balance Inquiry")
  print("4 for All Transactions")
  print("5 for Exit\n")

  ch = int(input("Enter Your Choice :"))

  if ch==1:
    amt = int(input("Enter the amount for deposit :"))
    balance+=amt
    transaction.append(f"Deposit of {amt}")
    print(f"Deposit Successfull of {amt}")
  elif ch==2:
    amt = int(input("Enter the amount you want to withdraw :"))
    if amt>balance:
      print("Insuffecient Balance")
    else:
      balance-=amt
      transaction.append(f"Withdraw of {amt}")
      print(f"Withdraw successfull of {amt}")
  elif ch==3:
    print(f"Account balance :{balance}")
  elif ch==4:
    if len(transaction)==0:
      print("No Transactions")
    else:
      for i in transaction:
        print(i)
  elif ch==5:
    print("Exiting...!")
    break
  else:
    print("Invalid Choice !!")
print("Thank you for using")
```

EVENT TRACKER SET

```python
entered_std = set()
rejected_std = set()

n = int(input("Enter Entry attempt: "))

for i in range(n):
  name = input("Enter student names: ").lower()
  name2 = ("").join((name.split(" ")))

  if name2 in entered_std:
    print(name, "already entered. entry rejected")
    rejected_std.add(name2)

  else:
    print(name, "Entry allowed")
    entered_std.add(name2)
print("Event details:")
print("\nTotal unique entry :",len(entered_std))
for j in entered_std:
  print(j)
print("\nTotal rejected entry :",len(rejected_std))
for k in rejected_std:
  print(k)


```

ROOM ALLOTMENT USING DICTIONARY

```python
rooms = {100:None,101:None,102:None,103:None,104:None,105:None,106:None,107:None,108:None,109:None}


print("Available rooms :")
for j in rooms:
  print(j,end=" ")

n = int(input("Enter Entry attempt: "))
for i in range(n):
  rmno = int(input("Enter the room number :"))
  if rooms[rmno] !=None:
    print("Room already alloted !!")
  else:
    name = input("Enter your name :")
    rooms[rmno] = name
    print(f"Room Alloted successfully to {name}")


delrom = int(input("Enter the room to vacate :"))
if rooms[delrom]==None or delrom not in rooms:
  print("Cannot vacate Room or room already vacate")
else:
  rooms[delrom] = None
  print(f"Room checkout Successfully !!")


print("All Booked rooms details :")
for j in rooms:
  if rooms[j]!=None :
    print(f"{j} : {rooms[j]}")

print("All Vacate rooms details :")
for j in rooms:
  if rooms[j]==None :
    print(f"{j} : {rooms[j]}")
print("Thank You !!!")
```

ATM PROGRAM WITH FUNCTION

```python
def login(pin, correct_pin=1674):
    return pin == correct_pin


def deposit(balance, transactions):
    amt = int(input("Enter the amount for deposit : "))
    balance += amt
    transactions.append(f"Deposit of {amt}")
    print(f"Deposit Successful of {amt}")
    return balance


def withdraw(balance, transactions):
    amt = int(input("Enter the amount you want to withdraw : "))

    if amt > balance:
        print("Insufficient Balance")
    else:
        balance -= amt
        transactions.append(f"Withdraw of {amt}")
        print(f"Withdraw Successful of {amt}")

    return balance


def history(transactions):
    if not transactions:
        print("No Transactions")
    else:
        for t in transactions:
            print(t)


def view_balance(balance):
    print(f"Account balance : {balance}")

balance = 0
transactions = []
approve = False

for cnt in range(3):
    print(f"Available chances : {3 - cnt}")
    code = int(input("Enter your pin : "))

    if login(code):
        approve = True
        break
    else:
        print("Wrong Pin")

while approve:
    print("\n1 for Deposit")
    print("2 for Withdraw")
    print("3 for Balance Inquiry")
    print("4 for All Transactions")
    print("5 for Exit\n")

    ch = int(input("Enter Your Choice : "))

    if ch == 1:
        balance = deposit(balance, transactions)

    elif ch == 2:
        balance = withdraw(balance, transactions)

    elif ch == 3:
        view_balance(balance)

    elif ch == 4:
        history(transactions)

    elif ch == 5:
        print("Exiting...!")
        break

    else:
        print("Invalid Choice !!")

print("Thank you for using")
```

LIBRARY MANAGEMENT

```python
books = {}
allotment = {}

id = 0
n = int(input("Enter the no of books for entry :"))
for i in range(n):
  bName = input("Enter the name of the book :")
  bAuthor = input("Enter the author of the book :")
  bCopy = int(input("Enter the no. of copies :"))
  temp = {"id":id,"Name":bName,"Author":bAuthor,"Available":bCopy}
  books[id] = temp
  id+=1
  print(f"{bName} added successfully !!")

print("Books available :")
for k in books:
  print("ID",k,"Name :",books[k]["Name"],"No of copies available :",books[k]["Available"])


allotId = 0
z = int(input("Enter the no. of book to allot :"))
for i in range(z):
  bId = int(input("Enter the book id to allot : "))
  if books[bId]['Available']>0:
    uName = input("Enter users name :")
    uNo = int(input("Enter users phNo:"))
    allotment[allotId] = {'BID':bId,"User":uName,"BName":books[bId]['Name'],"Contact":uNo}
    print(f"{bId} alloted to user successfully !!")
    books[bId]['Available']-=1
    allotId+=1
  else:
    print("Book not available")

print("Available books :")
for k in books:
  print("ID",k,"Name :",books[k]["Name"],"No of copies available :",books[k]["Available"])

print("All Allotments")
for j in allotment:
  print("Allot_ID",j,": Book id :",allotment[j]["BID"],"Alloted to :",allotment[j]["User"])
```

TRAFFIC MANAGEMENT

```python
import time
color = "RED"
print("Signal Color :",color)
while True:
  col = input("Enter the color of the signal(RED,YELLOW,GREEN) or 0 to exit:")
  if col=="0":
    break
  col = col.upper()
  if col not in ("RED,YELLOW,GREEN") or col==color:
    print("Invalid Signal Color or Cannot be the same as previous color!!")
    continue
  tim = int(input("Enter the time in second :"))
  while(tim>0):
    print(f"Time left {tim} second")
    time.sleep(1)
    tim-=1
  color = col
  print("Color changed to :",color)
```
